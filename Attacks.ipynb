```python
import pandas as pd 
import seaborn as sns 
import plotly.express as px
import matplotlib.pyplot as plt
import geopandas as gpd

ri = pd.read_csv("rba-dataset.csv")
ri
```

```python
# This is to find the Number of total attacks attempts 

Total_attack_attempts = ri[ri['Is Attack IP'] == True]
Total_attack_attempts
```

```python
#This is the actual number of number of attacks

Total_attack_attempts['Is Attack IP'].value_counts()
```

```python
Total_attack_attempts['Country'].value_counts()
```

```python
Total_attack_attempts['Region'].value_counts()
```

```python
#The number of successful attempts 
Total_success= ri[(ri['Login Successful'] == True) & (ri['Is Attack IP'] == True)]
Total_success
```

```python
Total_success['Browser Name and Version'].value_counts()
```

```python
Total_success['OS Name and Version'].value_counts()
```

```python
Total_success['Device Type'].value_counts()
```

```python
sns.countplot(
    data=Total_attack_attempts,
    x = 'Login Successful',
)

plt.title('Total Attack Attempts Unsuccessful and Successful of 104,231')
plt.xlabel('Login Successful')
plt.ylabel('Number of Attack Attempts')
plt.show()
```

# We can see that of the total successful attack attempts is 33,615 out of 104,231

```python
plt.figure(figsize=(20,6))
sns.countplot(
    data=Total_attack_attempts,
    x = 'Country',
)

plt.title('Number of Attack attempts by Country of all attack attempts 104,231')
plt.xlabel('Country')
plt.ylabel('Number of Attack Attempts')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,6))
sns.countplot(
    data=Total_success,
    x = 'Country',
)

plt.title('Number of Attack attempts by Country of Successful Attempts')
plt.xlabel('Country')
plt.ylabel('Number of Attack Attempts')
plt.xticks(rotation=90)
plt.show()
```
